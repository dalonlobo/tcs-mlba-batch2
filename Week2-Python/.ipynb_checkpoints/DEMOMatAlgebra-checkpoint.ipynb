```python
#Demonstrate matrix multiplication element wise and using numpy
# Find the time difference
import numpy as np
import time
def matmult(x, y):
    t1 = time.time()
    shape_x = x.shape
    shape_y = y.shape
    if shape_x[1] == shape_y[0]:
        z = np.zeros((shape_x[0], shape_y[1]), dtype=np.int64)
        for i in range(shape_x[0]):
            for j in range(shape_y[1]):
                c = 0
                for k in range(shape_y[0]):
                    c += x[i,k] * y[k,j]
                z[i,j] = c
        t2= time.time()
        print ("matmult::t2-t1" ,round(t2-t1, 5), "secs")
        #print ("z = ", z)
        return z
    else:
        print ("Dimensions should match. shape_x[1] <> shapey[0]", shape_x, shape_y)
        return None


def matmultNp(x, y):
    t1 = time.time()
    shape_x = x.shape
    shape_y = y.shape
    if shape_x[1] == shape_y[0]:
        z = np.dot(x, y)
        t2 = time.time()
        print ("matmultNP::", t2-t1, "secs")
        #print ("z = ", z)
        return z
    else:
        t2 = time.time()
        print("Dimensions should match. shape_x[1] <> shapey[0], time = ", shape_x, shape_y, t2 -t1)
        return None

def matAddNP(x, y):
    t1 = time.time()
    z = None
    if (x.shape == y.shape):
        z = x+y
    t2 = time.time()
    print ("matAddNP:", t2-t1, "secs")
    return z

def matAdd(x, y):
    t1 = time.time()
    z = np.zeros((x.shape))
    if (x.shape == y.shape):
        for i in range (x.shape[0]):
            for j in range(x.shape[1]):
                z [i,j ]= a[i,j] + b[i,j]
    t2 = time.time()
    print ("matAdd:", t2-t1, "secs")
    return z

def matDivideLC(x,y):
    assert (x.shape == y.shape)
    t1 = time.time()
    z = [[x[i,j]/y[i,j]] for i in range(x.shape[0]) for j in range(x.shape[1])]
    #z is a list (as this is created from a list comprehension

    z = np.array(z) #convert to np
    z = z.reshape(x.shape) #then reshape from (x.shape[0]*shape[1] to x.shape[0], x.shape[1]
    t2 = time.time()
    print("matDivideLC::", t2 - t1, "secs", z.shape)
    return z

def matDivide(x,y):
    assert (x.shape == y.shape)
    z = np.zeros((x.shape), dtype=np.float)
    t1 = time.time()
    for i in range(x.shape[0]):
        for j in range(x.shape[1]):
            z[i,j] = x[i,j]/y[i,j]
    t2 = time.time()
    print("matDivide::", t2 - t1, "secs", z.shape)
    return z

def matDivideNP(x,y):
    t1 =time.time()
    z =x/y
    t2 = time.time()
    print ("matDivideNP::", t2-t1, "secs")
    return z

#Check Mat multiply
a = np.arange(1, 10001, 1)
a = a.reshape((500, -1))
print ("a.shape after a.reshape(500, -1)=", a.shape)
#a = np.array([[1, 2], [3,4], [5,6]])
b = np.arange(-1, -10001, -1)
b = b.reshape((20, -1))
print ("b.shape after b.reshape((20, -1)=", b.shape)
c = matmult(a,b)
d = matmultNp(a,b)
print ("c.shape=", c.shape, "d.shape=", d.shape)
if ((np.sum(c == d) == d.shape[0]*d.shape[1])):
    print ("Same", np.sum(c==d))
else:
    print ("different")

#check mat Add
a = np.arange(1, 1000001, 1)
a = a.reshape((5000, -1))
b = np.arange(-1, -1000001, -1)
b = b.reshape((5000, -1))
c = matAdd(a,b)
d = matAddNP(a,b)
print ("c.shape=", c.shape, "d.shape=", d.shape)
if ((np.sum(c == d) == d.shape[0]*d.shape[1])):
    print ("Same", np.sum(c==d))
else:
    print ("different")

#check Mat Div
a = np.arange(1, 100001, 1)
a = a.reshape(5000, 20)

b = np.arange(-1, -100001, -1)
b = b.reshape(5000, -1)

c = matDivide(a,b)
c1 =matDivideLC(a,b)
c2 = matDivideNP(a,b)
print (np.sum(c ==c1), np.sum(c==c1) == a.shape[0]*a.shape[1])
#
# #if axis skipped sum over all numbers
# d = np.arange(1, 31, 1)
# d = d.reshape(3,5,2)
# print (d.sum())
```
