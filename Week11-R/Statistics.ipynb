## T-test ##

```r
gene_ALL = '283 752 517 313 590 355 243 568 732 512 851 234 613 436 433 319 1503 382 559 894 520 465 488 590 419 235 384'
gene_AML = '1544 484 787 534 1508 731 1217 254 440 621 1085'
```

```r
ALL_samples = as.numeric(unlist(strsplit(gene_ALL, split = " ")))
```

```r
AML_samples = as.numeric(unlist(strsplit(gene_AML, split = " ")))
```

```r
ALL_samples
AML_samples
```

```r
t.test(ALL_samples, AML_samples)
```
