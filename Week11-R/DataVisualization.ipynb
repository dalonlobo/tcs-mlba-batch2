## GGPLOT ##

```r
library(ggplot2)
```

```r
dataset_animal$'Skin Cover' <- as.factor(dataset_animal$'Skin Cover')
```

```r
class(dataset_animal$'Skin Cover')
```

```r
ggplot(dataset_animal, aes(x=dataset_animal$'Skin Cover'))+geom_bar( fill = 'pink')+ xlab("Skin Cover")
```

```r
data(mpg)
head(mpg)
```

```r
ggplot(mpg, aes(factor(cyl), cty))+geom_boxplot( aes(fill = factor(year) ) ) + xlab("Cylinders") + ylab("City Mileage") + labs(fill='Year of Manufacture') 
```

```r
base <- ggplot(mpg, aes(displ, cty))+geom_point()
```

```r
base + geom_smooth(aes(colour = "loess"), method = "loess", se = FALSE) + 
  geom_smooth(aes(colour = "lm"), method = "lm", se = FALSE) + labs(color = 'Smoothing Algorithm')
```

```r
ggplot(mpg, aes(cty))+geom_histogram(binwidth = 5, aes(fill=factor(cyl)))+labs(fill='No. of Cylinders')
```
